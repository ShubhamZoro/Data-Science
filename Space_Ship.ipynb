```python
df.year=df.year.astype(int)
success=df[df.Mission_Status=="Success"].sort_values(by=['year'])['year'].value_counts()
Failure=df[df.Mission_Status=="Failure"].sort_values(by=['year'])['year'].value_counts()
Partial_Failure=df[df.Mission_Status=="Partial Failure"].sort_values(by=['year'])['year'].value_counts()
Prelaunch_Failure=df[df.Mission_Status=="Prelaunch Failure"].sort_values(by=['year'])['year'].value_counts()

success.index
```

Comparison between success and failure over the year

```python
df.groupby(['Mission_Status'])['year'].count()
plt.figure(figsize=(15,6))
plt.scatter(success.index,success.values,color="green" ,linewidths = 2)
plt.scatter(Failure.index,Failure.values,color='blue',linewidths = 2)
plt.scatter(Partial_Failure.index,Partial_Failure.values,color='red',linewidths = 2)
plt.scatter(Prelaunch_Failure.index,Prelaunch_Failure.values,color='pink', linewidths = 2)
plt.xticks(rotation = 90)

plt.show()
```
